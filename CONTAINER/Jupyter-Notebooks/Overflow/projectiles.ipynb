# Projectile Motion

```python
red,lime,blue,green,purple = "#CF5970","#94B848","#878CC0","#5EBAA1","#BD5DC7"
```

## 2d Projectile, No Air Resistance

```python
# acceleration due to gravity
g = 32.174 # ft/s^2
```

```python
def get_x(v0,theta,t):
    ''' x motion solution.'''
    return v0 * cos(theta) * t

def get_y(v0,theta,t):
    ''' y motion solution.'''
    return v0 * sin(theta) * t - 0.5*g*t*t

def time_of_flight(v0,theta):
    ''' The total time of flight.'''
    return 2.0*v0*sin(theta)/g
```

### Get the solutions to the kinematical equations

```python
v0 = 78 # ft/s
theta = 38*2*pi/360 # rad
```

```python
t = linspace(0.0,time_of_flight(v0,theta),1000)
x = get_x(v0,theta,t)
y = get_y(v0,theta,t)
```

### Graph $x(t)$, $y(t)$ and $y(x)$

```python
figure(figsize=(10,5))
plot(t,x,'-', color=blue, linewidth=4)
xlabel('time (s)')
ylabel('$x$-distance (ft)')
title(r'$x(t)$ for $v_0 = $ %3.1f ft/s and $\theta =$ %3.1f rad$' % (v0,theta))
tight_layout()
savefig('x_vs_t.pdf')
```

```python
figure(figsize=(10,5))
plot(t,y,'-', color=green, linewidth=4)
xlabel('time (s)')
ylabel('$y$-distance (ft)')
title(r'$y(t)$ for $v_0 = $ %3.1f ft/s and $\theta =$ %3.1f rad$' % (v0,theta))
tight_layout()
savefig('y_vs_t.pdf')
```

```python
figure(figsize=(10,5))
plot(x,y,'-', color=red, linewidth=4)
xlabel('$x$-distance (ft)')
ylabel('$y$-distance (ft)')
title(r'$y(x)$ for $v_0 = $ %3.1f ft/s and $\theta =$ %3.1f rad$' % (v0,theta))
tight_layout()
savefig('y_vs_x.pdf')
```
