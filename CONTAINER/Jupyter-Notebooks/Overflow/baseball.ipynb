# Analysis of Projectile Motion in Baseball

```python
red,lime,blue,green,purple = "#CF5970","#94B848","#878CC0","#5EBAA1","#BD5DC7"
```

## Drag Coefficient

```python
def drag(v):
    ''' The drag coefficient of a baseball at sea level divided by mass in units of 1/feet'''
    vd = 35 * (2.23694) # mph
    Delta = 5 * (2.23694) # mph
    
    t1 = 0.0039
    t2 = 0.0058 / (1.0 + exp((v-vd)/Delta))
    return (t1 + t2)/(3.28084)
```

### Make a plot of the drag coefficient

```python
v = linspace(0,200,1000)
C = drag(v)

figure(figsize=(10,5))
plot(v,C,'-', color=blue, linewidth=4)
xlabel('Speed (mph)')
ylabel(r'$C \rho A / m$ (1/feet)')
#title(r'$x(t)$ for $v_0 = $ %3.1f ft/s and $\theta =$ %3.1f rad$' % (v0,theta))
tight_layout()
savefig('drag_coefficient.pdf')
```

```python
# acceleration due to gravity
g = 32.174 # ft/s^2

# Magnus drag
S0om = 4.1E-4
```

## The Equation of Motion for the Pitch
\begin{align}
a_x &= - \frac{C \rho A}{m} v_x v \\
a_y &= -g \\
a_z &= -\frac{S_0}{m} \omega v_x \\
\end{align}

```python
from scipy.integrate import odeint
```

```python
def eom(u,t,C0,omega):
    ''' The equation of motion (eom) for a pitched ball spinning around the y-axis.'''
    vx = u[0]
    vy = u[1]
    vz = u[2]
    x = u[3]
    y = u[4]
    z = u[5]
    v = sqrt(vx**2 + vy**2 + vz**2)
    return array([-C0*drag(v)*vx*v, -g, -S0om*omega*vx,vx,vy,vz])
```

### Setup the pitch

```python
t = linspace(0,0.5,100)
vx0,vy0,vz0,x0,y0,z0 = 134.9,0,0,0,6.5,0
init = [vx0,vy0,vz0,x0,y0,z0]
omega = 1900 * 2*pi / 60 # rad/s
args0 = (0.0,0.0)
argsF = (1.0,0.0)
argsC = (1.0,omega)
```

```python
omega
```

```python
# no dissipation or rotation
u0 = odeint(eom,init,t,args=args0)
r0 = u0[:,3:]

# fastball without spin
uF = odeint(eom,init,t,args=argsF)
rF = uF[:,3:]

# curveball
uC = odeint(eom,init,t,args=argsC)
rC = uC[:,3:]
```

### Plot the results

### x-y plane

```python
figure(figsize=(10,5))
axvline(x=59.9, color='k', linewidth=2)
plot(r0[:,0],r0[:,1],'-', color=blue, linewidth=4, label='No Spin, Vacuum')
plot(rF[:,0],rF[:,1],'-', color=red, linewidth=4, label='Fastball')
plot(rC[:,0],rC[:,1],'--', color=lime, linewidth=4, label='Curveball')
xlabel('$x$-distance (ft)')
ylabel('$y$-distance (ft)')
legend(loc='lower left', frameon=False, fontsize=20)
tight_layout()
savefig('pitch_xy.pdf')
```

### x-z plane

```python
figure(figsize=(10,5))
axvline(x=59.9, color='k', linewidth=2)
plot(r0[:,0],r0[:,2],'-', color=blue, linewidth=4, label='No Spin, Vacuum')
plot(rF[:,0],rF[:,2],'-', color=red, linewidth=4, label='Fastball')
plot(rC[:,0],rC[:,2],'--', color=lime, linewidth=4, label='Curveball')
xlabel('$x$-distance (ft)')
ylabel('$z$-distance (ft)')
legend(loc='lower left', frameon=False, fontsize=20)
tight_layout()
savefig('pitch_xz.pdf')
```
