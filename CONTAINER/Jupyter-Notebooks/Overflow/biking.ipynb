# Cycling at the Tour de France

## Define all the constant parameters of the model

```python
g,mu,m,rho,CA = 9.8,0.003,77,1.1,0.35
```

### Define the ODE integrator

\begin{align}
v[n] &=  v[n-1] + \Delta t \Big\{\frac{P}{m v[n-1]} - \frac{1}{2 m}C_D \rho A v[n-1]^2 - g (\mu_r \cos\theta + \sin\theta) \Big\} \\
x[n] &= x[n-1] + v[n]\Delta t \\
\end{align}



```python
def integrate(t0,x0,v0,dt,d,P,th):
    ''' Return the time, position and velocity arrays for a stage segment 
        at the Tour de France.'''

    # the initial conditions
    cx = [x0]
    cv = [v0]
    ct = [t0]
    n = 0
    while cx[n] <= d:
        n += 1
        # construct the time array
        ct.append(ct[n-1]+dt)
        
        # all the forces acting on the rider
        F_rider = P/(m*cv[n-1])
        F_air = (CA*rho/(2.0*m))*cv[n-1]**2
        F_g = g*sin(th)
        F_friction = mu*g*cos(th)
        
        # numerically integrate the equations of motion
        cv.append(cv[n-1] + dt*(F_rider - F_air - F_g - F_friction))
        cx.append(cx[n-1] + dt*cv[n])

    return ct,cx,cv
```

## Solve for the first part of the stage

```python
t,x,v = integrate(0.0,0.0,6.0,0.01,1500,325,2.0*pi*0.8/360)
```

### Graph the motion $v(t)$ and $x(t)$

```python
figure(figsize=(10,5))
plot(t,x,'-', color=blue, linewidth=4)
xlabel('time (s)')
ylabel('$x$ (m)')
tight_layout()
savefig('bike_x_vs_t.pdf')
```

```python
figure(figsize=(10,5))
plot(t,v,'-', color=green, linewidth=4)
xlabel('time (s)')
ylabel('$v$ (m/s)')
tight_layout()
savefig('bike_v_vs_t.pdf')
```

## Putting all the segments together

### The distances and heights

```python
d = [1500,2500,2000,3500,3000,3000]
y = [20,237,163,275,265,170]
```

### Compute the angles

```python
# store the angles in radians, but display in degrees
theta = [arcsin(1.0*y[n]/d[n]) for n in range(len(d))]

for n,th in enumerate(theta):
    print "theta_%d = %4.2f degrees" % (n+1,th*360/(2.0*pi))
```

### Define the angle dependent power function

```python
def power(theta):
    ''' The power output of an elite cyclist for different inclines.'''
    th = 360*theta/(2.0*pi)
    if th < 3.55:
        return 325.0
    elif theta < 5.16:
        return 425.0
    else:
        return 500.0
```

## Integrate over all stages to get the position and velocity and total stage time

```python
dt = 0.1 # s
t = []
x = []
v = []
t0 = 0.0 #s
x0 = 0.0 # m
v0 = 6.0 # m/s
print "%14s\t%14s\t%14s" % ("time (min)","distance (km)","speed (m/s)")
for n,cd in enumerate(d):
    ct,cx,cv = integrate(t0,x0,v0,dt,sum(d[:n+1]),power(theta[n]),theta[n])
    t.extend(ct)
    x.extend(cx)
    v.extend(cv)
    t0 = t[-1]
    x0 = x[-1]
    v0 = v[-1]
    print "%14.2f\t%14.2f\t%14.2f" % (t0/60,x0/1000,v0)
```
