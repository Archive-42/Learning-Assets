## Import the CSV datafile

```python
import pandas as pd
red,lime,blue,green,purple = "#CF5970","#94B848","#878CC0","#5EBAA1","#BD5DC7"
```

```python
gps = pd.read_csv('./gps_data.csv')
gps.head(10)
```

OK, this data format looks a little vague, let's see if we can understand what's going on here.

The only things we are really interested in are the **Timestamp** and the **Distance**.

The Timestamp and Distance values are not cumulative, but are recorded for a single interval.  To plot the data, we will need to generate the cumulative values.

## Turn TimeStamp into usable time data

We need to subtract of the inital *epoch* from each value

```python
time = []
t0 = gps['Timestamp'][0]
for ct in gps['Timestamp']:
    time.append(ct-t0)
time = array(time)
```

Now we have the cumulative time step in seconds.

```python
time
```

## Now for the distance, we need to sum up all values in the preceeding columns to get the running total, the individual intervals are in yards.

```python
distance = []
d = 0.0
for n,cd in enumerate(gps['Distance']):
    d += cd
    distance.append(d)
    #distance.append(sum(distance)+d)
distance = array(distance)
```

## Plot our distance vs. time graph

Remember, this is just distance along the path, it is not a vector

```python
figure(figsize=(10,5))
plot(time,distance,':o', color=blue, markersize=6, markeredgewidth=0.2, markeredgecolor='w')
xlabel('time (s)')
ylabel('distance (yards)')
title('Cumulative Distance During Walk Home')
```

## Compute the instantaneous velocity

Recall that the velocity is just the instantaneous slope on the position vs. time graph.  We can therefore compute the velocity using the individual data points through the formula:

$$ v = \frac{\Delta d}{\Delta t} $$

We will assume the initial velocity is zero

```python
velocity = zeros_like(time)
for n,cd in enumerate(distance):
    if n > 1:
        velocity[n] = (distance[n]-distance[n-1])/(time[n]-time[n-1])
```

Currently the units are such that the velocity is in yards per second.  This is quite awkward, let's convert everything to meters per second.

$$ 1 \frac{\text{yard}}{\text{s}} \times \frac{1\ \text{m}}{1.09361\ \text{yard}} = 0.9144 \frac{\text{m}}{\text{s}}$$

```python
velocity *= 1.0/1.09361
```

## Find the average velocity for the walk home

```python
ave_velocity = average(velocity)
print "The average velocity is v_av = %4.2f m/s" % ave_velocity
```

## Plot velocity vs time

```python
figure(figsize=(10,5))
axhline(y=ave_velocity, linewidth=0.8, color=green)
plot(time,velocity,':o', color=red, markersize=6, markeredgewidth=0.2, markeredgecolor='k')
xlabel('time (s)')
ylabel('speed (m/s)')
title('Instantaneous Velocity During Walk Home')
```

## Computing the acceleration

We can perform the same instantaneous slope measurment to get acceleration as a function of time, but it will likely be very noisy due to the large fluctuations.  Let's try. 

Now we will assume that the final acceleration is 0

```python
acceleration = zeros_like(time)
for n,cd in enumerate(velocity[:-1]):
    acceleration[n] = (velocity[n+1]-velocity[n])/(time[n+1]-time[n])
```

## Plot the acceleration vs time

```python
figure(figsize=(10,5))
plot(time,acceleration,':o', color=green, markersize=6, markeredgewidth=0.2, markeredgecolor='k')
xlabel('time (s)')
ylabel('acceleration (m/s$^2$)')
title('Acceleration During Walk Home')
```
