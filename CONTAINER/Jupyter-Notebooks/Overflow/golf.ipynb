# Analysis of Projectile Motion in Golf

```python
red,lime,blue,green,purple = "#CF5970","#94B848","#878CC0","#5EBAA1","#BD5DC7"
```

## Drag Coefficient

```python
def drag(v):
    ''' The drag coefficient of a golfball at sea level'''
    if v < 14.0:
        return 0.5
    else:
        return 7.0/v
```

### Make a plot of the drag coefficient

```python
v = linspace(0,70,200)
C = [drag(cv) for cv in v]

figure(figsize=(10,5))
plot(v,C,'-', color=blue, linewidth=4, label='Golf Ball')
axhline(0.5,color=red,linewidth=4, label='Smooth Ball')
xlabel('Speed (m/s)')
ylabel(r'$C$')
#title(r'$x(t)$ for $v_0 = $ %3.1f ft/s and $\theta =$ %3.1f rad$' % (v0,theta))
legend(frameon=False, fontsize=22)
ylim(0,1.0)
tight_layout()
savefig('golf_drag.pdf')
```
