# Can A Sprinter Run a Loop-the-Loop?

## Constants and Initial Conditions

```python
R = 3.0 # m
g = 9.8 # m/s^2
dt = 0.001 #s
f = 10 # N/kg
sigma = 0.9 #1/s
theta0 = 0.0 # rad
omega0 = 3.0 # rad/s
```

## The Equations of Motion
\begin{align}
\alpha &= \frac{1}{R}( f - \sigma R \omega - g \sin \theta) \\
\frac{F_N}{m g} &= \cos \theta + \frac{R \omega^2}{g}
\end{align}

```python
from scipy.integrate import odeint
```

```python
def eom(u,t,R):
    ''' The equation of motion (eom) for runner in plane polar coordinates.'''
    omega = u[0]
    theta = u[1]
    return array([(f - sigma*R*omega - g * sin(theta))/R,omega])
```

### Setup the attempt

```python
t = arange(0,5.0,dt)
init = [omega0,theta0]
```

### Integrate the equation of motion

```python
omega = []
theta = []
time = []
R = [2.5,3.0]
for n,cR in enumerate(R):
    co = []
    ctheta = []
    ct = []
    u = odeint(eom,init,t,args=tuple([cR]))
    index = where(u[:,1]< 2.0*pi)
    omega.append(u[index,0][0])
    theta.append(u[index,1][0])
    time.append(t[index])
```

### Plot the results

```python
figure(figsize=(10,5))
plot(time[0],theta[0]/(2.0*pi),'-', color=blue, linewidth=3, label=r'$R=%3.2f$ m'%R[0])
plot(time[1],theta[1]/(2.0*pi),'-', color=red, linewidth=3, label=r'$R=%3.2f$ m'%R[1])

ylabel(r'$\theta\, (2\pi)$')
xlabel(r'$t$ (s)')
legend(loc='upper left', frameon=False, fontsize=20)
tight_layout()
savefig('theta_vs_t.pdf')
```

```python
figure(figsize=(10,5))

axhline(y=sqrt(g*R[0]), linestyle=':',color=blue, linewidth=2, label=r'$v_{\mathrm{min}}(2.5)$')
plot(theta[0]/(2.0*pi),omega[0]*R[0],'-', color=blue, linewidth=3, label=r'$R=%3.2f$ m'%R[0])
axhline(y=sqrt(g*R[1]), linestyle=':',color=red, linewidth=2, label=r'$v_{\mathrm{min}}(3)$')
plot(theta[1]/(2.0*pi),omega[1]*R[1],'-', color=red, linewidth=3, label=r'$R=%3.2f$ m'%R[1])

xlabel(r'$\theta\, (2\pi)$')
ylabel('$v$ (m/s)')
legend(loc='upper left', frameon=False, fontsize=20, ncol=2)
tight_layout()
savefig('v_vs_theta.pdf')
```

### Plot the normal force

```python
figure(figsize=(10,5))

Nomg = cos(theta[0]) + R[0]*omega[0]**2/g
plot(theta[0]/(2.0*pi),Nomg,'-', color=blue, linewidth=3, label=r'$R=%3.2f$ m'%R[0])

Nomg = cos(theta[1]) + R[1]*omega[1]**2/g
plot(theta[1]/(2.0*pi),Nomg,'-', color=red, linewidth=3, label=r'$R=%3.2f$ m'%R[1])

xlabel(r'$\theta\, (2\pi)$')
ylabel('$F_N/m g$')
legend(loc='upper left', frameon=False, fontsize=20, ncol=1)
ylim(0,8)
tight_layout()
savefig('normal_vs_theta.pdf')
```
